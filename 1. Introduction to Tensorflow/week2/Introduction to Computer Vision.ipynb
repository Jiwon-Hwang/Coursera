### Fashion MNIST
* 70K Images
* 10 Categories
* Images are 28x28
* Can train a neural net!
> * 28x28로, grayscale로 정보 축소시켜도 충분히 분간 가능! (가방, 신발, 치마, 셔츠, 바지 등)  
> * 한 픽셀 당 0~255 값이므로 각각 1Byte씩! ==> 한 이미지 당 784 Byte만 필요

## Writing code to load training data

```
 # API를 통해 keras database에서 fashion mnist data 불러오기
 fashion_mnist = keras.datasets.fashion_mnist
 (train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data() # 4개 항목 return
```

## Coding a Computer Vision Neural Network
w0 x 0 + w1 x 1 + w2 x 2 ... wN x N = 9  
> * N : 128 (중간층 노드 수)
> * 9 : label (ex. 신발)

```
# 3개의 층으로 구성된 모델 구성 (입력층, 중간(은닉)층, 출력층)
model = keras.Sequential([
  keras.layers.Flatten(input_shape=(28, 28)),
  keras.layers.Dense(128, activation=tf.nn.relu),
  keras.layers.Dense(10, activation=tf.nn.softmax)
])
```
