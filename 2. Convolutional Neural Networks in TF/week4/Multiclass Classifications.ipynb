## Explore multi-class with Rock Paper Scissors dataset
#### [3가지 변경 사항들]
1. ImageDataGenerator 객체 생성 시, **class_mode**를 **'binary'**가 아니라 **'categorical'**로 지정해주어야 함! (**multi-class**)  

```py
train_datagen = ImageDataGenerator(rescale=1./255.) 

train_generator = train_datagen.flow_from_directory(
  train_dir,
  target_size=(300, 300),
  batch_size=128,
  class_mode='categorical')
```

2. Sequential 모델의 마지막 **출력층**인 Dense층을 **Dense(1, activation='sigmoid')**에서 **Dense(3, activation='softmax')**으로 바꿔줘야 함! (**multi-class**)
> cf. **softmax** : 여러 입력값을 받으며, 출력값들의 합이 1이 되게 하는 함수. 출력 범위는 모두 [0, 1] 사이.  
  
```py
model = tf.keras.models.Sequential([[
  ...
  tf.keras.layers.Dense(3, activation='softmax') # rock, paper, scissors
])
```

3. **compile** 시 **loss** 함수를 **'binary_crossentropy'**에서 **'categorical_crossentropy'**로 바꿔주면 됨.
  
```py
model.compile(loss='categorical_crossentropy',
    ...)
```

## Train a classifier with Rock Paper Scissors


## Test the Rock Paper Scissors classifier
paper 사진인데 네 개의 손가락이 딱 붙어있는 이미지 -> scissors로 잘못 분류하기도 함
